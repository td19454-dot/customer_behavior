```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
df.describe()
df.info()
df.isnull().sum()
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
df.isnull().sum()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
labels = ["young_adult","adult","middle_aged","senior"]
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head(10)
frequency = {'Fortnightly': 14,'Weekly' : 7, 'Monthly' : 30,'Quarterly':90,'Bi-Weekly':14,'Annually': 365,'Every 3 Months' : 90}
df['purchase_days']=df['frequency_of_purchases'].map(frequency)
df[['purchase_days','frequency_of_purchases']].head(20)
(df['discount_applied']==df['promo_code_used']).all()
df = df.drop('promo_code_used',axis = 1)
df.columns

df.to_csv('customer100.csv')
df.head(2000)
from sqlalchemy import create_engine

# MySQL connection
username = "root"
password = "trisha123"
host = "localhost"
port = "3306"
database = "customer_behavior"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Write DataFrame to MySQL
table_name = "customer"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)
pd.read_sql("SELECT * FROM customer LIMIT 20;", engine)from sqlalchemy import create_engine

# MySQL connection
username = "root"
password = "trisha123"
host = "localhost"
port = "3306"
database = "customer_behavior"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Write DataFrame to MySQL
table_name = "customer"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)
pd.read_sql("SELECT * FROM customer LIMIT 20;", engine)
```
